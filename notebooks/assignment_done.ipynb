Question 2: Write Python code to list the first 10 objects (blob name) in the "gcp-public-data-landsat" bucket, along with their sizes.

```python
#Write Python code to list the first 10 objects (blob name) in the "gcp-public-data-landsat" bucket, along with their sizes.

from google.cloud import storage

client = storage.Client()
bucket = client.get_bucket("gcp-public-data-landsat")

blobs = bucket.list_blobs()
count = 0
for blob in blobs:
    print(f"Name: {blob.name}, Size: {blob.size} bytes")
    count += 1
    if count >= 10:
        break
```
